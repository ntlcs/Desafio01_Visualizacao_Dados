```python
df_pt = pd.read_csv("student-por.csv", sep=";")
df_mat = pd.read_csv("student-mat.csv", sep=";")
```

Insight 1: Distribuição de Idades

```python
plt.figure(figsize=(10, 6))
plt.hist(df_pt['age'], bins=30, color='skyblue', edgecolor='black', alpha=0.7, label='Portuguese')
plt.hist(df_mat['age'], bins=30, color='orange', edgecolor='black', alpha=0.7, label='Math')
plt.title('Distribuição de Idades nos Datasets')
plt.xlabel('Idade')
plt.ylabel('Frequência')
plt.legend()
plt.show()
```

Insight 2: Relação entre Duas Variáveis Numéricas

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_pt['G1'], df_pt['G2'], color='skyblue', alpha=0.7, label='Portuguese')
plt.scatter(df_mat['G1'], df_mat['G2'], color='orange', alpha=0.7, label='Math')
plt.title('Relação entre G1 e G2 nos Datasets')
plt.xlabel('G1')
plt.ylabel('G2')
plt.legend()
plt.show()
```

Insight 3: Contagem de Categorias em uma Variável Categórica (usando Seaborn)

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='sex', data=df_pt, palette='viridis', label='Portuguese', alpha=0.7)
sns.countplot(x='sex', data=df_mat, palette='muted', label='Math', alpha=0.7)
plt.title('Contagem de Gêneros nos Datasets')
plt.xlabel('Gênero')
plt.ylabel('Contagem')
plt.legend()
plt.show()
```
